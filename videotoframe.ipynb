# video To frame Conversion for training data


```python
import os
import cv2
mydir=r"C:\Users\NEERAJ SINGH\Desktop\Major_project\videodataset\train"
path=os.walk(mydir)
filelist=[]
for(path,folders,files)in path:
    filepaths=[
        os.path.join(path,file)
        for file in files
        
    ]
    filelist.extend(filepaths)
for file in filelist:
    print(file)
    cap=cv2.VideoCapture(file)
    ret, image=cap.read()
    count=0
    while True:
        if ret==True:
                x=file.split('.')
                paths=x[0]+"%d.jpg"
                print(paths)
                cv2.imwrite( paths % count, image)
                cap.set(cv2.CAP_PROP_POS_MSEC,(count**10))
                ret,img=cap.read()
                count+=1
                print("saving frames")

        else:
            break



                
           
   
    
       
```
